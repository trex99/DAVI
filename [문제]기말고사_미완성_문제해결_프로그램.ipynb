```python
# z-score Í≥ÑÏÇ∞
score_price_original = data.price.values
score_price_mean = np.mean(score_price_original)
score_price_std = np.std(score_price_original)

print('ÌèâÍ∑† : ', score_price_mean)
print('ÌëúÏ§ÄÌé∏Ï∞® : ', score_price_std)
print('')

# ÌëúÏ§ÄÏ†êÏàòÌôîÌïú Îç∞Ïù¥ÌÑ∞ (ÌèâÍ∑†ÏóêÏÑúÏùò Îñ®Ïñ¥ÏßÑ Ï†ïÎèÑ)
z_score_price = (score_price_original - score_price_mean) / score_price_std
print('ÌëúÏ§ÄÏ†êÏàò Ìï©Í≥Ñ : ', z_score_price.sum())
print('ÌëúÏ§ÄÏ†êÏàò ÌèâÍ∑† : ', z_score_price.mean())
print('ÌëúÏ§ÄÏ†êÏàò ÌëúÏ§ÄÌé∏Ï∞® : ', z_score_price.std())
```
