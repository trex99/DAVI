## 데이터 이해

- Z-Score (표준점수) = ${(X - \mu) \over \sigma} $
  - $\mu$ : 표본 평균
  - $\sigma$ : 표본 표준편차

```python
import numpy as np

# 우리가 일반적으로 사용하는 데이터
score_math_X = np.array([70, 80, 90, 95])
score_math_mean = np.mean(score_math_X)
score_math_std = np.std(score_math_X)

print('평균 : ', score_math_mean)
print('표준편차 : ', score_math_std)
print('')

# 표준점수화한 데이터 (평균에서의 떨어진 정도)
z_score_math = (score_math_X - score_math_mean) / score_math_std
print('표준점수 : ', z_score_math)
print('표준점수 합계 : ', z_score_math.sum())
print('표준점수 평균 : ', z_score_math.mean())
print('표준점수 표준편차 : ', z_score_math.std())
```

### 이항분포
- 연속된 n번의 독립시행에서 각 시행이 성공(실패)할 확률 p를 가질 때 만들어지는 이산 확률 분포
- 이항분포를 따르면, 다음의 확률질량함수(pmf)로 확률 계산  
$$ P(X=k) = \frac{n!}{k! (n-k)!} p^{k}(1-p)^{n-k}$$
- 예) 15명의 환자가 치료될 확률이 0.7일 경우, 5명의 환자가 치료될 확률
- k = 5
- n = 15
- p = 0.7  
$$ P(X=5) = \frac{15!}{5! (15-5)!} 0.7^{5}(1-0.7)^{15-5} = 0.0029802869$$
- 예) 동전을 100회 던져서 앞면이 50회 나올 확률
- k = 50
- n = 100
- p = 0.5  
$$ P(X=20) = \frac{100!}{50! (100-50)!} 0.5^{50}(1-0.5)^{100-50} = 0.0795892373$$



### 이항분포 시각화

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def bin_dist(k, n, p):
    nck = np.math.factorial(n) / (np.math.factorial(k) * np.math.factorial(n - k))
    value = nck * p**k * (1-p)**(n-k)
    return value

# 15명의 환자가 치료될 확률이 0.7일 경우, 1~15명의 환자가 치료될 각각의 확률
k = None # 프로그램 반복으로 값이 설정, 여기서는 불필요
n = 15
p = 0.7

# 시각화
x = np.arange(n+1)
prob = np.array([bin_dist(k, n, p) for k in range(n+1)])
plt.title('n, p = 15, 0.3')
sns.barplot(x=x, y=prob, color='blue')
plt.show()
```

### 기하분포

- 연속된 n번의 독립시행에서 각 시행이 성공(실패)할 확률 p를 가질 때, 처음 성공까지 총 시도한 횟수($k$)가 만들어지는 이산 확률 분포
- 기하분포를 따르면, 다음의 확률 질량 함수로 확률 계산  
$$ P(X=k) = (1-p)^{k}p \;\;\; k=0,1,2,3....$$
$$ P(X=k) = (1-p)^{k-1}p \;\;\; k=1,2,3....$$
- 예) ○○소속의 야구선수가 안타칠 확률이 0.05라고 한다. 6타석만에 안타칠 확률
- k = 6
- p = 0.05
$$ P(X=6) = (1-0.05)^{6-1}0.05 = 0.0386890468 $$
- 예) ○○자동차전문학원은 0.25라고 소문이 났다. 적어도 3번 <font color=yellow>이내</font> 합격할 확률
- k = 1, k = 2, k = 3
- p = 0.25
- 이내 : 누적을 의미
$$ P(X=1) = (1-0.25)^{1-1}0.25 = 0.25 $$
$$ P(X=2) = (1-0.25)^{2-1}0.25 = 0.1875 $$
$$ P(X=3) = (1-0.25)^{3-1}0.25 = 0.140625 $$
- 누적확률 : 0.578125

### 기하분포 시각화

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def geo_dist(k, p):
    value = (1-p)**(k-1)*p
    return value

# 야구선수가 안타칠 확률이 0.05라고 한다. 1~20타석 각각의 안타칠 확률
k = 20
p = 0.05

x = np.arange(k+1)
prob = np.array([geo_dist(k, p) for k in range(k+1)])
plt.title('p = 0.7')
sns.barplot(x=x, y=prob, color='blue')
plt.show()
```

※ 점추정량은 구간추정의 기본통계량으로 사용

### 구간추정

```python
# 무작위 숫자가 적힌 주머니에서 일부를 꺼내서 구간추정
import scipy.stats as st
import numpy as np

circle_sample = [5, 10, 17, 29, 14, 25, 16, 13, 9, 17]
data = len(circle_sample) - 1
mu = np.mean(circle_sample) # 표본 평균
se = st.sem(circle_sample) # 표준 오차

# 95% 신뢰구간 : 확률적으로 계산된 범위내 95%가 포함될것으로 예상
result_1 = st.t.interval(0.95, data, mu, se)

# 99% 신뢰구간 : 확률적으로 계산된 범위내 99%가 포함될것으로 예상
result_2 = st.t.interval(0.99, data, mu, se)

print('95% 신뢰구간 : ', result_1)
print('99% 신뢰구간 : ', result_2)
```

```python
# 신뢰구간 표시로 의미있는 구간추정 시각화
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np

circle_sample = [5, 10, 17, 29, 14, 25, 16, 13, 9, 17]
mu = np.mean(circle_sample)
se = st.sem(circle_sample)

# 표준 오차를 반영하여 구한 정규분포표
norm_dist = st.norm(scale = se)

# 95% 신뢰 구간 -> 0.975 (1-0.95=0.05, 0.05/2=0.025)
err = norm_dist.ppf(0.975)

plt.bar(['circle_sample'], [mu], color = ['orange'], yerr = [err], alpha = 0.6) # capsize = 10
plt.show()
```

## 가설 검정

- 1단계 : 귀무가설($H_0$)와 대립가설($H_1$) 설정
  - 귀무가설($H_0$) : 직접 검정의 대상이 되는 가설, 기각해야만 하는 가설
  - 대립가설($H_1$) : 귀무가설이 기각될 때 받아들여지는 가설, 내가 증명하고 싶은 가설
- 2단계 : 유의수준 α 결정
- 3단계 : 검정통계량과 p-value 계산
  - p-value : 어떤 사건이 우연히 발생할 확률 (보통 5%보다 크면 우연 발생으로 보고 의미가 없다고 판단)
- 4단계 : 기각역 판정
- 5단계 : 검정통계량이 기각역 안에 들어가면 귀무가설 H0 기각 반대면 귀무가설 H1 채택

### 단일표본 T-검정

- 하나의 집단에 대한 특성 중 평균에 대한 가설검정
- 관심 대상이 되는 변수가 양적변수(quantitative variable)인 경우 적용

○○전자 회사의 웹 사이트는 온라인 자동차 부품을 판매하고 있다.
웹 사이트 관리자는 자동차 부품을 구매하기 위한 방문자가 웹 페이지에 오래 접속하면 구매로 이어진다는 사실을 알았다.
그래서 웹 사이트 관리자는 평균 접속시간(session time)이 길어야 160초 이고 이를 늘리기 위해서는 서버 교체 사업이 필요하다고 주장한다.

❓ 과연 웹 사이트 관리자는 평균 접속시간이 길어야 160초라고 주장하였다. 이 주장이 얼마나 믿을만 한가?

- 1단계 : 귀무가설($H_0$)와 대립가설($H_1$) 설정
  - $H_0: \mu=160 \mbox{ (or } \mu \geq 160)$
  - $H_1: \mu < 160$
- 2단계 : 유의수준 α 결정
  - α = 0.05
- 3단계 : 검정통계량과 p-value 계산
$$ \mbox{T-검정통계량} =\frac{\bar x - \mu_0 }{s/\sqrt n } \sim t_{n-1}$$


```python
# 데이터 불러오기
import pandas as pd
import numpy as np

session_times = pd.read_csv('web_page_data.csv')
session_times.head(3)
```

- 점 추정 : 표본 평균

```python
# 한 집단에 대한 평균 추론
# 점 추정 : 표본평균 계산
s_mean = session_times.sTime.mean()
print('표본평균 : ', s_mean)
```

- 구간 추정 : 표본 평균에 대한 신뢰구간

```python
# 방법 #1
from statsmodels.stats import weightstats

stat = weightstats.DescrStatsW(session_times.sTime)
s_result = stat.tconfint_mean(alpha=0.05)
print('95% 신뢰구간 : ', s_result)
```

```python
# 방법 #2
# 무작위 숫자가 적힌 주머니에서 일부를 꺼내서 구간추정하는 프로그램 코드를 사용해서 검증
import scipy.stats as st
import numpy as np

circle_sample = session_times.sTime  # 공 데이터를 접속시간 데이터로 대체
data = len(circle_sample) - 1
mu = np.mean(circle_sample) # 표본 평균
se = st.sem(circle_sample) # 표준 오차

# 95% 신뢰구간 : 확률적으로 계산된 범위내 95%가 포함될것으로 예상
result_1 = st.t.interval(0.95, data, mu, se)
print('95% 신뢰구간 : ', result_1)
```

- t-검정 실시

```python
# 방법 #1
from scipy import stats

t_stat, p_value = stats.ttest_1samp(session_times.sTime, 160)
print(f'T_statistics: {t_stat:.3f}, p-value: {p_value:.3f}')
```

```python
# 방법 #2
from statsmodels.stats import weightstats

stat = weightstats.DescrStatsW(session_times.sTime)
t_result = stat.ttest_mean(value=160)
print(f'T_statistics: {t_result[0]:.3f}, p-value: {t_result[1]:.3f}')
```

- 4단계 : 기각역 판정
  - p-value = 0.551 > α(=0.05)
  - 유의수준 0.05에서 귀무가설을 기각하지 못함
  - 따라서 평균 접속시간이 160초보다 작다는 충분한 통계적 근거가 없음
- 5단계 : 검정통계량이 기각역 안에 들어가면 귀무가설 H0 기각, 들어가지 않으면 귀무가설 H1 채택
  - 결론 : 평균 접속시간은 160초 이상이기 때문에 서버 교체 사업은 필요하지 않다.
