### 데이타 사전 지식

YouTube(세계적으로 유명한 동영상 공유 웹사이트)는 플랫폼에서 가장 인기 있는 동영상 목록을 관리합니다. 

버라이어티 매거진에 따르면 올해의 최고 인기 동영상을 결정하기 위해 YouTube는 사용자 상호 작용(조회 수, 공유 수, 댓글 및 좋아요 수)을 측정하는 등 여러 요인을 조합하여 사용합니다. 해당 연도 전체에서 가장 많이 본 동영상은 아닙니다.

YouTube 트렌딩 목록의 최고 성능은 뮤직 비디오(유명하게 남성적인 "Gangam Style" 등), 유명인 및/또는 리얼리티 TV 공연, YouTube에서 잘 알려진 무작위 카메라 바이럴 동영상입니다.

분석대상 데이터 셋은 인기 급상승 YouTube 동영상의 일일 기록입니다.
