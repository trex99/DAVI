# 시각화 용어

- figure : 그림이 담길 틀
- figsize : 그림이 담길 틀의 크기
- ax : 하나의 그림 (axes)
- plot : 선 그래프 (bar, boxplot, hist, scatter...)
- lim : 범위 (xlim, ylim)
- title : plot 전체 이름
- label : 축 이름 (xlabel, ylabel)
- grid : 그림의 일정 간격의 가로, 세로 눈금
- axis : 가로 또는 세로축
- tick : 그래프 축에 간격을 구분하기 위한 표시 눈금
- marker, dot : 그래프 상의 특정 값 표시
- legend : 범례

## 예제 - 3 (한글 및 기호문제)

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_path = 'C:/Windows/Fonts/NGULIM.TTF'
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rcParams['font.family'] = font_name
plt.rcParams['font.size'] = 10
plt.rcParams['axes.unicode_minus'] = False
```
